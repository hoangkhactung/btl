````
import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from datasets import load_dataset
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
STOPWORDS = set(stopwords.words('english'))

import spacy
nlp = spacy.load("en_core_web_sm", disable=["parser","ner"])
````

````
# tải dataset từ HuggingFace
ds = load_dataset("UniverseTBD/arxiv-abstracts-large", split="train")
df = ds.to_pandas()

# hàm tách categories
def parse_categories(cat):
    if not isinstance(cat, str):
        return []
    parts = re.split(r'[;,|\s]+', cat.strip())
    return [p for p in parts if p]

df["categories_list"] = df["categories"].fillna("").astype(str).apply(parse_categories)
df["primary_category"] = df["categories_list"].apply(lambda x: x[0] if len(x) > 0 else "unknown")

# lấy top 10 categories
topk = 10
top_cats = df["primary_category"].value_counts().nlargest(topk).index.tolist()
df_filtered = df[df["primary_category"].isin(top_cats)].copy()

print("Dataset sau khi lọc:", df_filtered.shape)
df_filtered[["title","primary_category"]].head()
````

````
# phân bố nhãn
plt.figure(figsize=(10,5))
sns.countplot(y="primary_category", data=df_filtered,
              order=df_filtered["primary_category"].value_counts().index)
plt.title("Phân bố nhãn (Top 10 categories)")
plt.show()

# histogram độ dài abstract
df_filtered["abstract_len"] = df_filtered["abstract"].fillna("").astype(str).apply(lambda x: len(x.split()))
plt.figure(figsize=(8,4))
sns.histplot(df_filtered["abstract_len"], bins=80, kde=False)
plt.title("Histogram độ dài abstract")
plt.xlabel("Số từ")
plt.ylabel("Số bài báo")
plt.show()
````

````
from wordcloud import WordCloud

# ghép text từ 5000 abstract ngẫu nhiên
text_all = " ".join(df_filtered['abstract'].fillna("").astype(str).values[:5000])

wc = WordCloud(
    width=1200, height=600,
    background_color='white',
    stopwords=STOPWORDS
).generate(text_all)

plt.figure(figsize=(12,6))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.title("Wordcloud abstracts (sample 5000)")
plt.show()
````

````
# Hàm tiền xử lý
def clean_text(text):
    if not isinstance(text, str):
        return ""
    s = text.lower()                                # chuyển về chữ thường
    s = re.sub(r'http\S+|www\S+|https\S+', ' ', s)  # bỏ link
    s = re.sub(r'\d+', ' ', s)                      # bỏ số
    s = re.sub(r'[^\w\s]', ' ', s)                  # bỏ dấu câu, ký tự đặc biệt
    s = re.sub(r'\s+', ' ', s).strip()              # bỏ khoảng trắng thừa
    return s

# Áp dụng tiền xử lý
df_filtered["clean_abstract"] = df_filtered["abstract"].fillna("").astype(str).apply(clean_text)

# Xem thử kết quả
df_filtered[["abstract","clean_abstract"]].head(5)
````

````
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
import re
from datasets import load_dataset

# Load lại dataset (chỉ lấy 10k mẫu cho nhẹ)
ds = load_dataset("UniverseTBD/arxiv-abstracts-large", split="train[:10000]")
df = ds.to_pandas()

# Parse categories
def parse_categories(cat):
    if not isinstance(cat, str): return []
    return re.split(r'[;,|\s]+', str(cat).strip())

df["categories_list"] = df["categories"].fillna("").astype(str).apply(parse_categories)
df["primary_category"] = df["categories_list"].apply(lambda x: x[0] if len(x)>0 else "unknown")

# Lấy top 10 categories
topk = 10
top_cats = df["primary_category"].value_counts().nlargest(topk).index.tolist()
df_filtered = df[df["primary_category"].isin(top_cats)].copy()

print("✅ df_filtered:", df_filtered.shape)

# Vector hóa TF-IDF
tfidf = TfidfVectorizer(
    max_features=20000,
    stop_words="english",
    ngram_range=(1,2)
)

X = tfidf.fit_transform(df_filtered["abstract"].fillna("").astype(str))
y = df_filtered["primary_category"]

# Chia train/test
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)

print("✅ TF-IDF xong.")
print("Train shape:", X_train.shape, " Test shape:", X_test.shape)
````

````
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score

# Train Logistic Regression
clf = LogisticRegression(max_iter=1000, class_weight="balanced", n_jobs=-1)
clf.fit(X_train, y_train)

# Dự đoán trên tập test
y_pred = clf.predict(X_test)

# Kết quả
print("📌 Classification Report:")
print(classification_report(y_test, y_pred))
print("✅ Accuracy:", accuracy_score(y_test, y_pred))
````

````
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Tính confusion matrix
cm = confusion_matrix(y_test, y_pred, labels=clf.classes_)

# Vẽ heatmap
plt.figure(figsize=(10,7))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=clf.classes_,
            yticklabels=clf.classes_)
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix (TF-IDF + Logistic Regression)")
plt.xticks(rotation=45, ha="right")
plt.show()
````

````
import numpy as np

# Lấy danh sách feature (từ TF-IDF)
feature_names = np.array(tfidf.get_feature_names_out())

# In top 10 từ quan trọng nhất cho mỗi nhãn
for i, class_label in enumerate(clf.classes_):
    top10 = np.argsort(clf.coef_[i])[-10:][::-1]  # lấy 10 từ có hệ số cao nhất
    print(f"\n📌 Class: {class_label}")
    print(", ".join(feature_names[top10]))
````

````
# 📝 Kết luận và thảo luận

## 1. Kết quả đạt được
- Đã xây dựng pipeline **phân loại chủ đề bài báo khoa học (arxiv abstracts)**.
- Tiến hành các bước:
  - Chuẩn bị & tiền xử lý dữ liệu.
  - Phân tích dữ liệu (EDA) với biểu đồ phân bố nhãn, histogram độ dài abstract, wordcloud từ phổ biến.
  - Vector hóa văn bản bằng **TF-IDF** (uni-gram và bi-gram).
  - Huấn luyện mô hình **Logistic Regression**.
  - Đánh giá bằng **Accuracy, Precision, Recall, F1** và trực quan hóa bằng **Confusion Matrix**.
  - Trích xuất **top keywords cho mỗi nhãn** để diễn giải mô hình.

- Với tập dữ liệu 10k abstract (top 10 categories), mô hình Logistic Regression đạt:
  - Accuracy ≈ XX% (thay bằng kết quả thật của mày).
  - F1-score trung bình ≈ YY% (thay bằng kết quả thật).
  - Confusion matrix cho thấy mô hình nhầm lẫn nhiều ở các nhãn có từ vựng trùng lặp.

## 2. Hạn chế
- Dữ liệu còn **mất cân bằng** giữa các nhãn.
- TF-IDF chưa nắm bắt được ngữ nghĩa sâu (chỉ dựa trên tần suất từ).
- Dữ liệu gốc rất lớn (2.3 triệu abstract) nhưng chỉ dùng được một phần nhỏ do hạn chế bộ nhớ Colab.

## 3. Hướng phát triển
- Thử nghiệm các phương pháp vector hóa hiện đại hơn:
  - **Word embeddings** (Word2Vec, GloVe).
  - **Sentence embeddings** (SBERT, SciBERT).
- Huấn luyện mô hình mạnh hơn: **SVM, XGBoost, Neural Networks**.
- Thử **fine-tuning BERT/SciBERT** trên toàn bộ dataset để cải thiện kết quả.
- Tối ưu tiền xử lý (lemmatization, loại bỏ stopwords, bigram/trigram).
- Xây dựng ứng dụng demo bằng **Streamlit** cho phép nhập abstract và dự đoán category ngay.
````
