````
import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from datasets import load_dataset
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
STOPWORDS = set(stopwords.words('english'))

import spacy
nlp = spacy.load("en_core_web_sm", disable=["parser","ner"])
````

````
# ti dataset t HuggingFace
ds = load_dataset("UniverseTBD/arxiv-abstracts-large", split="train")
df = ds.to_pandas()

# hm tch categories
def parse_categories(cat):
    if not isinstance(cat, str):
        return []
    parts = re.split(r'[;,|\s]+', cat.strip())
    return [p for p in parts if p]

df["categories_list"] = df["categories"].fillna("").astype(str).apply(parse_categories)
df["primary_category"] = df["categories_list"].apply(lambda x: x[0] if len(x) > 0 else "unknown")

# ly top 10 categories
topk = 10
top_cats = df["primary_category"].value_counts().nlargest(topk).index.tolist()
df_filtered = df[df["primary_category"].isin(top_cats)].copy()

print("Dataset sau khi lc:", df_filtered.shape)
df_filtered[["title","primary_category"]].head()
````

````
# phn b nhn
plt.figure(figsize=(10,5))
sns.countplot(y="primary_category", data=df_filtered,
              order=df_filtered["primary_category"].value_counts().index)
plt.title("Phn b nhn (Top 10 categories)")
plt.show()

# histogram  di abstract
df_filtered["abstract_len"] = df_filtered["abstract"].fillna("").astype(str).apply(lambda x: len(x.split()))
plt.figure(figsize=(8,4))
sns.histplot(df_filtered["abstract_len"], bins=80, kde=False)
plt.title("Histogram  di abstract")
plt.xlabel("S t")
plt.ylabel("S bi bo")
plt.show()
````

````
from wordcloud import WordCloud

# ghp text t 5000 abstract ngu nhin
text_all = " ".join(df_filtered['abstract'].fillna("").astype(str).values[:5000])

wc = WordCloud(
    width=1200, height=600,
    background_color='white',
    stopwords=STOPWORDS
).generate(text_all)

plt.figure(figsize=(12,6))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.title("Wordcloud abstracts (sample 5000)")
plt.show()
````

````
# Hm tin x l
def clean_text(text):
    if not isinstance(text, str):
        return ""
    s = text.lower()                                # chuyn v ch thng
    s = re.sub(r'http\S+|www\S+|https\S+', ' ', s)  # b link
    s = re.sub(r'\d+', ' ', s)                      # b s
    s = re.sub(r'[^\w\s]', ' ', s)                  # b du cu, k t c bit
    s = re.sub(r'\s+', ' ', s).strip()              # b khong trng tha
    return s

# p dng tin x l
df_filtered["clean_abstract"] = df_filtered["abstract"].fillna("").astype(str).apply(clean_text)

# Xem th kt qu
df_filtered[["abstract","clean_abstract"]].head(5)
````

````
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
import re
from datasets import load_dataset

# Load li dataset (ch ly 10k mu cho nh)
ds = load_dataset("UniverseTBD/arxiv-abstracts-large", split="train[:10000]")
df = ds.to_pandas()

# Parse categories
def parse_categories(cat):
    if not isinstance(cat, str): return []
    return re.split(r'[;,|\s]+', str(cat).strip())

df["categories_list"] = df["categories"].fillna("").astype(str).apply(parse_categories)
df["primary_category"] = df["categories_list"].apply(lambda x: x[0] if len(x)>0 else "unknown")

# Ly top 10 categories
topk = 10
top_cats = df["primary_category"].value_counts().nlargest(topk).index.tolist()
df_filtered = df[df["primary_category"].isin(top_cats)].copy()

print(" df_filtered:", df_filtered.shape)

# Vector ha TF-IDF
tfidf = TfidfVectorizer(
    max_features=20000,
    stop_words="english",
    ngram_range=(1,2)
)

X = tfidf.fit_transform(df_filtered["abstract"].fillna("").astype(str))
y = df_filtered["primary_category"]

# Chia train/test
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)

print(" TF-IDF xong.")
print("Train shape:", X_train.shape, " Test shape:", X_test.shape)
````

````
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score

# Train Logistic Regression
clf = LogisticRegression(max_iter=1000, class_weight="balanced", n_jobs=-1)
clf.fit(X_train, y_train)

# D on trn tp test
y_pred = clf.predict(X_test)

# Kt qu
print(" Classification Report:")
print(classification_report(y_test, y_pred))
print(" Accuracy:", accuracy_score(y_test, y_pred))
````

````
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Tnh confusion matrix
cm = confusion_matrix(y_test, y_pred, labels=clf.classes_)

# V heatmap
plt.figure(figsize=(10,7))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=clf.classes_,
            yticklabels=clf.classes_)
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix (TF-IDF + Logistic Regression)")
plt.xticks(rotation=45, ha="right")
plt.show()
````

````
import numpy as np

# Ly danh sch feature (t TF-IDF)
feature_names = np.array(tfidf.get_feature_names_out())

# In top 10 t quan trng nht cho mi nhn
for i, class_label in enumerate(clf.classes_):
    top10 = np.argsort(clf.coef_[i])[-10:][::-1]  # ly 10 t c h s cao nht
    print(f"\n Class: {class_label}")
    print(", ".join(feature_names[top10]))
````
